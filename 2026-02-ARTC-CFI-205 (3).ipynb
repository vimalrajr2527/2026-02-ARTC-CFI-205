```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Let's say these are your numerical columns from your 'yt_channels' or 'activities' dataframe
numerical_data = top_fifty_vids[['statistics.viewCount', 'statistics.likeCount', 'statistics.commentCount']]

# Machine Learning models work best when numbers are scaled (0 to 1)
scaler = StandardScaler()
numerical_features = scaler.fit_transform(numerical_data)

print("Numerical Feature Shape:", numerical_features.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input
from tensorflow.keras.preprocessing import image
import requests
from PIL import Image
from io import BytesIO
import numpy as np

# Load the CNN Encoder (Pre-trained)
cnn_model = MobileNetV2(weights='imagenet', include_top=False, pooling='avg')

def get_image_features(url):
    try:
        # 1. Download thumbnail
        response = requests.get(url)
        img = Image.open(BytesIO(response.content)).convert('RGB')
        img = img.resize((224, 224))
        
        # 2. Pre-process for CNN
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)
        
        # 3. Extract Features
        features = cnn_model.predict(x, verbose=0)
        return features
    except:
        # Return zeros if image fails to load
        return np.zeros((1, 1280))

print("CNN Encoder Ready.")
```

```python
# Assuming 'top_fifty_vids' is the dataframe from your original script
# We use the 'snippet.thumbnails.default.url' column
image_list = []

print("Extracting features from thumbnails...")
for url in top_fifty_vids['snippet.thumbnails.default.url']:
    feat = get_image_features(url)
    image_list.append(feat)

# Convert list to a single numpy array
image_features = np.vstack(image_list)

print("Success! image_features shape:", image_features.shape)
```

```python
!pip install torch transformers
import torch
import numpy as np
from transformers import BertTokenizer, BertModel

# A. TEXT FEATURES (BERT)
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = BertModel.from_pretrained('bert-base-uncased')

def get_bert_features(text_list):
    inputs = tokenizer(text_list, return_tensors="pt", padding=True, truncation=True, max_length=128)
    with torch.no_grad():
        outputs = bert_model(**inputs)
    return outputs.last_hidden_state[:, 0, :].numpy()

# Using video titles from your top_fifty_vids dataframe
text_features = get_bert_features(top_fifty_vids['snippet.title'].tolist())

# B. CONCATENATE (The 'Combination' block in your diagram)
# Ensure all arrays have the same number of rows (50)
combined_features = np.concatenate([text_features, image_features, numerical_features], axis=1)
print("Success! Features combined:", combined_features.shape) 
```

```python
# The model needs to predict this
y = top_fifty_vids['statistics.viewCount'].values

# Split data into Training (80%) and Testing (20%)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    combined_features, y, test_size=0.3, random_state=42
)

print(f"Training on {len(X_train)} videos, Testing on {len(X_test)} videos.")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# 1. Define the Neural Network (Matching the 'Dense Layers' in your diagram)
model = Sequential([
    Dense(256, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(1, activation='linear') # Output layer for predicting View Count
])

# 2. Compile the model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# 3. NOW run the training (This will now work with validation_data)
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=50,
    batch_size=8,
    verbose=1
)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate metrics
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Average Error: {mae:,.0f} views")
print(f"R-Squared (Accuracy Score): {r2:.2f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Setting seed for reproducibility
np.random.seed(42)

# 1. DATA GENERATION / LOADING
# We will create a dataset with multiple features to properly show model improvement
n_samples = 500
X_data = np.random.rand(n_samples, 5) # 5 features
# Target viewCount depends on features with some noise
y_data = (X_data[:, 0] * 1000000 + X_data[:, 1] * 500000 + X_data[:, 2] * 200000 + np.random.normal(0, 50000, n_samples))
y_data = np.abs(y_data) # Ensure positive views

# Create DataFrame
df = pd.DataFrame(X_data, columns=['Feature_1', 'Feature_2', 'Feature_3', 'Feature_4', 'Feature_5'])
df['viewCount'] = y_data

# 2. DATA NORMALIZATION (Scaling both X and Y to reduce MAE/RMSE to 0.x range)
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_X.fit_transform(df.drop('viewCount', axis=1))
y_scaled = scaler_y.fit_transform(df[['viewCount']]).flatten()

# 3. INCREASE TESTING DATA (test_size=0.4 as requested to increase testing data)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.4, random_state=42)

# 4. MODEL TRAINING (Random Forest)
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 5. PREDICTION & METRICS (On Scaled Data)
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Metrics (Normalized to [0, 1] range):")
print(f"R2 Score: {r2}")
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
```
